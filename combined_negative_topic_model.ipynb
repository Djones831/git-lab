```python
import pandas as pd
from bertopic import BERTopic


reddit_df = pd.read_parquet("reddit_with_vader.parquet")
bluesky_df = pd.read_parquet("bluesky_with_vader.parquet")
```

```python
reddit_df["origin"] = "reddit"
bluesky_df["origin"] = "bluesky"
```

```python
neg_reddit = reddit_df[reddit_df['sentiment_label'] == "negative"]
neg_bluesky = bluesky_df[bluesky_df['sentiment_label'] == "negative"]
```

```python
neg_reddit.shape
```

```python
neg_bluesky.shape
```

```python
# Downscaling neg_reddit into the same size as bluesky via random sampling
neg_reddit = neg_reddit.sample(n=neg_bluesky.shape[0], random_state=42)
```

```python
combined_neg = pd.concat([neg_reddit, neg_bluesky])
```

```python
docs = combined_neg['clean_post'].tolist()

topic_model = BERTopic(
    language="english",
    calculate_probabilities=True,
    verbose=True
)

topics, probs = topic_model.fit_transform(docs)
```
